```
#LOGISTIC
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics

from IPython.display import display, HTML

dataset = pd.read_csv('data.csv')

#Redundant variables
del dataset["id"]
del dataset["age_days"]

'''dataset = dataset[dataset.height <= 200]
dataset = dataset[dataset.height >= 120]
dataset = dataset[dataset.weight <= 160]
dataset = dataset[dataset.ap_hi.between(0,500)]
dataset = dataset[dataset.ap_lo.between(0,2000)]'''
'''
del dataset["gender"]
del dataset["alco"]'''

#separating the x and y variables
display(HTML(dataset.head(10).to_html()))
X = dataset.iloc[:, :-1]
print(X)
y = dataset.iloc[:, -1]
print(y)

#spliting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1)

#scaling the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)

#Building the model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 1)
classifier.fit(X_train, y_train)

#predicting the testing data
y_pred = classifier.predict(X_test)
print(y_pred)
print(y_test)
#print(np.concatenate((y_pred, y_test),1))

#creating confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))

```

```
#per label classification
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
